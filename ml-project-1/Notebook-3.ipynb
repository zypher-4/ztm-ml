## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y= [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 5. Save & show
fig.savefig("images/sample-plot.png")
```

### Two oprions from subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 10))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.randint(10, size=(15,1)), np.random.randint(100, size=(15,1)))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

ax1.set(title='Graph 1',
        xlabel='X Axis',
        ylabel='Y Axis')
ax2.set(title='Graph 2',
        xlabel='X Axis',
        ylabel='Y Axis')
ax3.set(title='Graph 3',
        xlabel='X Axis',
        ylabel='Y Axis')
ax4.set(title='Graph 4',
        xlabel='X Axis',
        ylabel='Y Axis')

plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
#Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randint(10, size=(15,1)), np.random.randint(100, size=(15,1)))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

ax[0, 0].set(title='Graph 1')

plt.show()
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something advanced, use the OO method

```python
over_50 = hd[hd['age'] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO method mixed with pyplot
fig, ax =plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
ax.set_xlim([45, 100])
plt.show()
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

## Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

## Customize the plot
ax.set(title='Heart Disease and Cholestrol Levels',
       xlabel='Age',
       ylabel='Cholestrol')

## Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax.axhline(y=over_50['chol'].mean(),
           linestyle='--')

## Show plot
plt.show()

## Save the figure

fig.savefig('images/Heart-Disease-and-Cholestrol-Levels.png')
```

```python
## OO method from scratch
fig, ((ax1), (ax2)) = plt.subplots(nrows=2,
                                   figsize=(10, 12))

## Plot the data
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

## Customize the plot
ax1.set(title='Heart Disease and Cholestrol Levels',
       xlabel='Age',
       ylabel='Cholestrol')

## Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax1.axhline(y=over_50['chol'].mean(),
           linestyle='--')

## Plot the data
scatter = ax2.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])

## Customize the plot
ax2.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

## Add a legend
ax2.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax2.axhline(y=over_50['thalach'].mean(),
           linestyle='--')

## Show plot
plt.show()
```

```python
## OO method from scratch
fig, ((ax1), (ax2)) = plt.subplots(nrows=2,
                                   ncols=1,
                                   figsize=(10, 12),
                                   sharex=True)

## Plot the data
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

## Customize the plot
ax1.set(title='Heart Disease and Cholestrol Levels',
       # xlabel='Age',
       ylabel='Cholestrol')

## Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax1.axhline(y=over_50['chol'].mean(),
           linestyle='--')

## Plot the data
scatter = ax2.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])

## Customize the plot
ax2.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

## Add a legend
ax2.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax2.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

## Add a title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
## OO method from scratch
fig, ((ax1), (ax2)) = plt.subplots(nrows=2,
                                   ncols=1,
                                   figsize=(10, 10),
                                   sharex=True)

## Plot the data
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

## Customize the plot
ax1.set(title='Heart Disease and Cholestrol Levels',
       # xlabel='Age',
       ylabel='Cholestrol')

# Change the x axis limit
ax1.set_xlim([50,80])

## Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax1.axhline(y=over_50['chol'].mean(),
           linestyle='--')

## Plot the data
scatter = ax2.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='summer')

## Customize the plot
ax2.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

# Change the y axis limit
ax2.set_ylim([60,200])

## Add a legend
ax2.legend(*scatter.legend_elements(), title='Target')

## Add a horizontal line
ax2.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

## Add a title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');

fig.savefig('images/final.png')
```
