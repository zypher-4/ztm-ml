## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y= [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 5. Save & show
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.cos(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Dan\'s Nut Butter Store',
       ylabel='Prices ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# Make some data for histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

### Two oprions from subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 10))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.randint(10, size=(15,1)), np.random.randint(100, size=(15,1)))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

ax1.set(title='Graph 1',
        xlabel='X Axis',
        ylabel='Y Axis')
ax2.set(title='Graph 2',
        xlabel='X Axis',
        ylabel='Y Axis')
ax3.set(title='Graph 3',
        xlabel='X Axis',
        ylabel='Y Axis')
ax4.set(title='Graph 4',
        xlabel='X Axis',
        ylabel='Y Axis')

plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
#Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randint(10, size=(15,1)), np.random.randint(100, size=(15,1)))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

ax[0, 0].set(title='Graph 1')

plt.show()
```
